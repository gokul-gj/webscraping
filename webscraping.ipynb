```python
soup=BeautifulSoup(data.text)
```

```python
books=soup.select('ol.row')[0]#for selecting css selection
```

```python
links=books.find_all('a')#selecting all the tags
```

```python
links=[l.get('href') for l in links]
```

```python
links_new=[]
for i in links:
    if i not in links_new:
        links_new.append(i)
```

```python
for j in range(1,len(full_links)):
    book_link=full_links[j]
    print(full_links[j])
    data1=requests.get(book_link)
    soup1=BeautifulSoup(data1.text)
    rating=((soup1.select('article p')[2]['class'][1]))
    print(rating)
    name=soup1.select('h1')
    print(name)
    name=(str(name))[5:-6]
    catergory=re.split(r'[< >]',str(soup1.find_all('a')[3]))
    catergory=catergory[3]
    details=pd.read_html(data1.text,match="UPC")
    df_new_1=details[0]
    dic={'url':book_link,'title':name,'review_rating':rating,'category':catergory}
    df_11=pd.DataFrame(dic.items())
    df_new_1=df_new_1.append(df_11,ignore_index=True)
    df_new=pd.concat([df_new, df_new_1[1]], axis=1)
```
